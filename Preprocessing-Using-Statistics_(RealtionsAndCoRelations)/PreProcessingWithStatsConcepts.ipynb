```python
from scipy.stats import pearsonr
#pearson is a coorelation body
```

```python
from scipy.stats import pearsonr

correlation = {
    feature: pearsonr(df_edit[feature], df_edit['Conservation Status'])[0]
    for feature in list_cols
}
#pearsonr return r-value and p-value ,by takin pearson(x,y) [0] it will return only r-value
```

```python
correlation.items()
```

```python
corr_df = pd.DataFrame(list(correlation.items()) , columns=['Feature' , 'R-Value'])
```

```python
corr_df.sort_values(by='R-Value' , ascending=False)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

alpha = 0.05

# Convert Conservation Status into categorical bins
df_edit['Conservation Status Bins'] = pd.qcut(df_edit['Conservation Status'], q=4, labels=False,duplicates="drop")

chi2_results = {}

# Run chi2 test for each categorical feature
for feature in list_cols:
    # Build contingency table
    contingency = pd.crosstab(df_edit[feature], df_edit['Conservation Status Bins'])
    
    # Perform Chi-Square test
    chi2_stat, p_val, dof, expected = chi2_contingency(contingency)
    
    # Save results
    decision = (
        'Reject Null → Dependent (Keep Feature)' 
        if p_val < alpha 
        else 'Accept Null → Independent (Drop Feature)'
    )
    
    chi2_results[feature] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'degrees_of_freedom': dof,
        'Decision': decision
    }

# Convert results to DataFrame for better readability
chi2_df = pd.DataFrame(chi2_results).T
print(chi2_df)
```
