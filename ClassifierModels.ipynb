```python
import pandas as pd
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
#Logistic Regression Implementation
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(['deck','embark_town' , 'alive','class','who','adult_male'] , axis=1 , inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
#Fill blank ages and  and removing embarked null rows

df['age'].fillna(df['age'].mean(), inplace=True)
df.dropna(subset=['embarked'] , inplace=True)
```

```python
df.info()
```

```python
df2 = df.copy()
```

```python
# we have to predict survived col
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
```

```python
df2['sex'] = le.fit_transform(df['sex'])
df2['embarked'] = le.fit_transform(df['embarked'])
```

```python
df2.head()
```

```python
df2.astype(int)
```

```python
#No need for scaling as we are simply implementing the Logistic regression
```

```python
x = df2.drop('survived' , axis=1)
y = df2['survived']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
model = LogisticRegression()
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
# now we have to calculate the accuracy of the model for that we'll be using confusion metrics
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
asc = accuracy_score(y_test , y_pred)
```

```python
asc
```

```python
confusion_matrix(y_test , y_pred)
```

```python
classification_report(y_test , y_pred)
```
